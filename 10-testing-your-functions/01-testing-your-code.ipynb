## Installing pytest with pip

Even though Python includes a lot of functionality, we've already seeen that third-party packages offer useful tools, such as `numpy`.
Sometimes, third-party packages are adopted into the standard library, although this generally happens with packages that are unlikely to change going forward.

Many packages are kept out of the standard library so they can be developed at their own rate. This allows for more frequent updates and additions than if they were tied to Python's development schedule. Take full advantage of these third-party packages! They are incredibly useful.

### Updating pip

We've already seen an example of using the `pip` tool to install `numpy`. Now that we're more familiar with the langauge, let's take a deeper dive. First, we'll update pip to make sure any known security issues have been addressed.

Open a terminal through VS Code and issue the following command:
```bash
$ python -m pip install --upgrade pip
```
The first part of this command, `python -m pip`, tells Python to run the module `pip`. This is necessary when upgrading `pip` on Windows, where there are limitations about programs modifying themselves. In general, accessing `pip` through Python is the more robust route, since you're forced to specify the versin of Python you'd like to use.

The second part, `install --upgrade`, tells `pip` to updated a package that's already installed. The last part, `pip`, specifies which third-party package should be updated.

You can use this command to update any third-party package, for example, let's see if there are any updates to `numpy`:
```bash
$ python -m pip install --upgrade numpy
```

### Installing pytest

Next, let's install `pytest`:
```bash
$ python -m pip install pytest
```
Because we're running in a virtual environment, we can ommit any flags such as `--user`, since all packages installed in this environment will only be accessible to the current user (you!) running this particular environment.
