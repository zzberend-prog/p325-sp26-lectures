## Higher-Dimiensional Arrays

### Matrices and Arrays

One dimensional arrays are useful for representing vectors, or lists of data. When mathematically modeling a system, you will also need two dimensional arrays for representing matrices, or tables of data.

A matrix of numbers is organized in columns and rows. For example:
$$
\left[
\begin{matrix}
0 & 12 & -1 & 5 \\
-1 & -1 & -1 & 0 \\
11 & 5 & 5 & -2
\end{matrix}
\right]
$$

The above is an example of a $3 \times 4$ matrix, i.e., a table with 3 rows and 4 columns.

If we associate the symbol $A$ with a matrix, $A_{i,j}$ denotes the number in row number $i$ and column number $j$. Counting rows and columns from 0, the above matrix has $A_{0,0} = 0$ and $A_{2,3}=-2$.

Matrices can be added and subtracted in the same piecwise way as vectors. They can can also be multiplied or divided by a scalar.

We can generalize the concept of vector or matrix to *array*, which holds quantities with $d$ indices, or has *rank* $d$. For $d=3$, an array $A$ has elements $A_{p,q,r}$. If $p$ goes from $0$ to $n_p-1$, $q$ from $0$ to $n_q-1$, and $r$ from $0$ to $n_r-1$, the array has $n_p \times n_q \times n_r$ total elements. The shape of this array is the tuple $(n_p, n_q, n_r)$.

### Creating Two-Dimensional Arrays

To make a two dimensional NumPy array, you can start by making a table out of nested lists:

```python
# Two lists of temperatures in Celcius and Fahrenheit
Cdegrees = [ -30 + i*10 for i in range (3) ]
Fdegrees = [ 9/5*C + 32 for C in Cdegrees ]

# Nested lists make a table
table = [ [C, F] for C, F in zip(Cdegrees, Fdegrees) ]
print(table)
```

Similarly to lists and 1-dimensional arrays, you can now turn this table into a 2-dimensional array:

```python
import numpy as np

# Make a matrix from nested lists
matrix = np.array(table)
print(matrix)
```

We can also use the `np.zeros` function, passing a shape instead of a simple length:

```python
matrix0 = np.zeros((5,5))
print(matrix0)
```

### Accessing Two-Dimensional Arrays

You can access the elements of this matrix by in the same way you would access a list of lists, first specifying the outer list (row) and then the inner list (column):

```python
# Access element in row 2, column 1
print(matrix[2][1])
```

For arrays, you will often see a different but equivalent synax:

```python
# Access element in row 2, column 1
print(matrix[2,1])
```

Like we saw before, we can inspect the shape of this matrix using the `shape` attribute:

```python
# Print shape information
print(matrix.shape)
```

In the case of our temperature matrix, it has 3 rows and 2 columns.

We can use this to loop over all elements of the matrix using two `for` loops:

```python
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        print(matrix[i,j])
```

### Slicing Two-Dimensional Arrays

Arrays also allow for slicing each dimension in the same way you slice lists. If we wanted to extract the celcius degrees in column 1, we could write:

```python
# Extract all rows of column 1
print(matrix[0:matrix.shape[0], 1])

# More compactly
print(matrix[:, 1])
```

Let's make a larger array to explore this further:

```python
bigmatrix = np.linspace(1,30,30).reshape(5,6)
print(bigmatrix)
```

If we wanted to make a slice of the rows 3 and above, for columns 2 to 4, we could write:

```python
print(bigmatrix[3:, 2:5])
```

Here, we started by specify `3:`, which picked out the rows starting with `19.` and `25.`. We then specified `2:5` which began with the column starting with `3.` and went to (but didn't include) the column starting with `5.`.
