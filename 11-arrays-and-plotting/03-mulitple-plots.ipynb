# Arrays and Plotting

## Ploting Multiple Curves

### Multiple Curves on the Same Plot

You will often want to plot two curves on the same axes to compare them. For example, you might want to show that the envelope of the damped harmonic oscillator follows an exponential decay. To do this, you can simply call `plt.plot()` multiple times.

```python
import matplotlib.pyplot as plt
import numpy as np

def damped(t, beta=0.5):
    return np.exp(-beta*t)

def damped_oscillator(t, T=1, beta=0.5):
    return np.exp(-beta*t)*np.cos(2*np.pi*t/T)

t_values = np.linspace(0,10,1000)
y_values = damped_oscillator(t_values)
e_values = damped(t_values)

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(t_values, y_values, label=r'$e^{-\beta t}\cos{\frac{2 \pi}{T}}$')
ax.plot(t_values, e_values, label=r'$e^{-\beta t}$', linestyle=':')

# Set chart title and label axes
ax.set_title('Damped harmonic oscillator', fontsize=24)
ax.set_xlabel('Amplitude', fontsize=14)
ax.set_ylabel('Time [s]', fontsize=14)

# Set size of tick labels
ax.tick_params(labelsize=14)

plt.legend()
```

### Multiple Plots in the Same Figure

It is also convenient to include separate plots in the same figure. For example, if we want to investigate how changing the period parameter $T$ affects the oscillator, we could simply draw both:

```python
import matplotlib.pyplot as plt
import numpy as np

def damped_oscillator(t, T=1, beta=0.5):
    return np.exp(-beta*t)*np.cos(2*np.pi*t/T)

t_values = np.linspace(0,10,1000)
y1_values = damped_oscillator(t_values)
y2_values = damped_oscillator(t_values, T=2)

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.plot(t_values, y1_values, label='Damped Oscillator (T=1)', c='blue')
ax.plot(t_values, y2_values, label='Damped Oscillator (T=2)', c='red')

# Set chart title and label axes
ax.set_title('Damped harmonic oscillator', fontsize=24)
ax.set_xlabel('Amplitude', fontsize=14)
ax.set_ylabel('Time [s]', fontsize=14)

# Set size of tick labels
ax.tick_params(labelsize=14)

plt.legend()
```

Because oscillating functions with different periods tend to look a bit messy when plotted together, you could instead plot then next to each other:

```python
import matplotlib.pyplot as plt
import numpy as np

def damped_oscillator(t, T=1, beta=0.5):
    return np.exp(-beta*t)*np.cos(2*np.pi*t/T)

t_values = np.linspace(0,10,1000)
y1_values = damped_oscillator(t_values)
y2_values = damped_oscillator(t_values, T=2)

plt.style.use('seaborn-v0_8')
fig, axs = plt.subplots(2, 1)
axs[0].plot(t_values, y1_values, label='Damped Oscillator (T=1)', c='blue')
axs[0].set_ylabel('Amplitude [m]', fontsize=14)
axs[0].tick_params(labelsize=14)

axs[0].legend()

axs[1].plot(t_values, y2_values, label='Damped Oscillator (T=2)', c='red')
axs[1].set_ylabel('Amplitude [m]', fontsize=14)
axs[1].set_xlabel('Time [s]', fontsize=14)
axs[1].tick_params(labelsize=14)

axs[1].legend()
```
