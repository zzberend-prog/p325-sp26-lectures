# Random Numbers

## Drawing Random Integers

It is often the case that you need to draw random integers instead of floating point numbers, e.g., a value among 1, 2, 3, and 4.
One way to do this is to draw a real number from the uniform distribution between 0 and 1 and divide this interval into four equal sub-intervals:

```python
import random

r = random.random()
if r < 0.25:
    r = 1
elif r < 0.5:
    r = 2
elif r < 0.75:
    r = 3
else:
    r = 4
```

### Random Integer Functions

Because this is such a common task, there are built in functions within the `random` module that do this.
If you want to draw an integer in $[a,b]$, you can use the `randint()` function:

```python
import random

print(random.randint(1,4))
```

Alternatively, with the NumbPy module this looks like:

```python
import numpy as np

rng = np.random.default_rng()
print(rng.integers(1,5,size=5))
```

Notice that the NumPy version of the function requires that you put $b+1$ in for the upper limit.
Since we wanted the numbers 1 to 4, we passed an upper limit of 5.

### Example: Throwing a dice

Let's make a function that lets the computer throw a die `N` times and returns the fraction of the throws the die shows six eyes:

```python
import numpy as np

def six_eyes(N):
    """Throw a die N times and return the fraction of times a six is shown"""
    rng = np.random.default_rng()

    eyes = rng.integers(1,7,size=N)
    success = eyes == 6        # True/False array
    n_sixes = np.sum(success)  # treats True as 1, False as 0

    return n_sixes/N

print(f'After throwing 1000 die, the fraction of sixes is {six_eyes(1000)}')
```

The `eyes == 6` conditional statement does an element-wise comparison and results in an array with `True` or `False`, depending on whether that element is `6` or not.
When this boolean array is passed to `np.sum()` it treats `True` as if it were 1 and `False` as if it were 0, so the sum is the number of elements in `eyes` that equals 6.
Because this code uses the vectorized NumPy functions, it runs very efficiently.

#### Vectorized Version with Batches

One feature of vectorized code is that it must store the entire list of random numbers in memory.
Depending on how long your array is, and how much memory your computer has, this can be a problem.

Instead of drawing all of the random numbers at once, we could instead draw them in batches of `array_size`.
This will result in `N//array_size` batches of the same size, plus a remainder batch.
Note the double slash in `N//array_size`, which means integer division (what we want here).

We can now update the program to the following:

```python
import numpy as np

def six_eyes(N, array_size=1_000_000):
    """Throw a die N times and return the fraction of times a six is shown"""
    # Split all experiments into batches of array_size, plus a rest batch, then
    # make a list containing the size of each batch
    n_batches = N//array_size
    rest_size = N%array_size
    batch_sizes = [array_size]*n_batches + [rest_size] 

    rng = np.random.default_rng()

    n_sixes = 0
    for batch_size in batch_sizes:
        eyes = rng.integers(1,7,size=batch_size)
        success = eyes == 6         # True/False array
        n_sixes += np.sum(success)  # treats True as 1, False as 0

    return n_sixes/N

print(f'After throwing 12.34 million die, the fraction of sixes is {six_eyes(int(1.234e7))}')
```

### Drawing a Random Element from a List

Since array indices are integers, we can also use these tools to draw a random element from an array:

```python
import random

numbers = list(range(2,101,2))                   # list of even numbers, 2 ... 100
re = numbers[random.randint(0, len(numbers)-1)]  # pick a random element of the list
print(re)
```

There is also a function to choose a random element:

```python
import random

numbers = list(range(2,101,2))
re = random.choice(numbers)
print(re)
```

Alternatively, you can randomly shuffle the elements in a list:

```python
import random

numbers = list(range(2,11,2))
random.shuffle(numbers)
print(numbers)
```

The `random` module in NumPy has the same functions.
