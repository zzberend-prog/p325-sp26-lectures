## Computing Probabilities

You've likely studied how to compute probabilities in one or more of your math courses.

This section will discuss how you can pivot to numerical methods for computing these probabilities.

### Monte Carlo Simulation

Assume that we perform $N$ experiments where the outcome is random.
If some event takes place $M$ times in these $N$ experiments, an estimate of the probability is $M/N$.
This estimate becomes more accurate as $N$ increases, and we assume the exact probability is given as $N \rightarrow \infty$.

Programs that run a large number of experiments and record the outcome are called *simulations*.
The mathematical technique of performing a large number of experiments on a computer based on drawing random numbers is commonly called *Monte Carlo simulation*.

In particle physics, the rules of quantum mechanics (and its relativistic extension quantum field theory) are probabilistic.
For a given proton–proton collision, the particles that come out the other side are a random process that depends on the particle type, energy and momentum, and other quantum numbers.
Another example is offshore engineering, where environmental loads from wind, currents, and waves show random behaviors.

Python provides a straightforward way to run Monte Carlo simulations of random physical phenomenon.
However, as systems become more complicated, compiled languages like C++ become necessary to reduce the amount of time the program takes to run.
Using vectorization is one way to balance the ease-of-use of Python with the numerical efficiency of C++, since the vectorized functions often call compiled C++ code under the hood.

### Example: Throwing Dice

Consider throwing two dice, one black and one green.
What is the probability that the number of eyes on the black die is larger than the number of eyes on the green die?

#### Straightforward Solution

Let's simulate `N` throws of the two dice in a program.
After each throw, we can check if the black die has more eyes than the green die and increment $M$ if so.

```python
import random

N = 1000
M = 0

for _ in range(N):
    black = random.randint(1,6)
    green = random.randint(1,6)
    if black > green:
        M += 1

print(f'The probability that the black die shows more eyes than the green die is {M/N}')
```

#### Vectorized Solution

Let's practice a vectorized solution, even if this particular example isn't incredibly taxing for our computers.
The main difference will be requesting an array of random numbers and operating on it with vectorized functions.
Essentially, we'll avoid writing any explicit `for` loops.

```python
import numpy as np

# Number of experiments to perform
N = 1_000_000

# Generate experiments in a 2xN array
rng = np.random.default_rng()
data = rng.integers(1,7, size=(2,N))

# Slice out the results for the black and green die separately
black = data[0,:]
green = data[1,:]

# Calculate probability
success = black > green
M = np.sum(success)
print(f'The probability that the black die shows more eyes than the green die is {M/N}')
```

#### A game

Instead of a probability, let's play a game where you pay $1 to throw these two die.
You win $2 if there are more eyes on the black than the green die.
Is this game fair?

We can modify the above program to find out.

```python
import numpy as np

# Initial parameters
N = 1_000_000
initial_money = 10

# Generate experiments in a 2xN array
rng = np.random.default_rng()
data = rng.integers(1,7, size=(2,N))

# Slice out the results for the black and green die separately
black = data[0,:]
green = data[1,:]

# Calculate profits
money = -1*N  # You paid $1 for every game

success = black > green
M = np.sum(success)
money += 2*M      # You won  $2 when there were more black eyes than green eyes

print(f'After {N} games, your average earning per game is ${money/N}')
```
