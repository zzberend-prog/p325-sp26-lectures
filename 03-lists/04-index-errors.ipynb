## Avoiding Index Errors When Working with Lists

One error you'll commonly see is an *index error*.

```python
quarks = ['down', 'anti-down'] # pion
print(quarks[3])
```

Python attempts to give you the item at index `3`. But when it searches the list, no item in `quarks` has an index of `3`. Because of the off-by-one nature of indexing in lists, this error is very common.

An index error means Python can't find an item at the index you requested. If you ever get this error, try decreasing the index you're asking for by one, then run the prgorma again to see if you get the correct results.

Keep in mind that whenever you want to access the last item in a list you should use the index `-1`. This will always work, even if you have changed your list since the last time you accessed it.`

```python
quarks = ['beauty', 'strange']
print(quarks[-1])

quarks.append('down')
print(quarks[-1])
```

The only time this approach will cause an error is when you request the last item from an empty list:

```python
quarks = []
print(quarks[-1])
```

If an index error occurs and you can't figure out how to resolve it, try printing you list or just printing the length of your list. It may look different than you're expecting!
