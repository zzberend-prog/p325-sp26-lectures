## Using if Statements with Lists

When combining lists and `if` statements, you can watch for special values that need to be treated differently than other values in the list. You can also efficiently manage changing conditions.

### Checking That a List Is Not Empty

Let's go back to the pizza example. Say we want to print all the requested ingredients. However, we want to protect against someone accidentally requesting a boring pizza. You can accomplish this with:

```python
requested_toppings = ['anchovies']

if requested_toppings:
    for topping in requested_toppings:
        print(f'Adding {topping}.')
else:
    print("Are you sure you want a plain pizza?")
```

When the name of a list is used in an `if` statement, Python returns `True` if the list contains at least one item; an empty list evaluates to `False`. Try adding a topping to the list above to check that it behaves as expected.

### Using Multiple Lists

We should also protect against customers asking for a pizza topping that isn't available. To do this, we can make two lists: one with the requsted toppings and one with the available toppings. Before adding a topping to the pizza, we can check if the requested topping is available.

```python
available_toppings = ['mushrooms', 'olives', 'green peppers',
                      'pepperoni', 'pineapple', 'extra cheese']

requested_toppings = ['mushrooms', 'french fries', 'extra cheese']

for requested_topping in requested_toppings:
    if requested_topping in available_toppings:
        print(f'Adding {requested_topping}.')
    else:
        print(f"Sorry, we don't have any {requested_topping}.")

print("\nFinished making your pizza!")
```
