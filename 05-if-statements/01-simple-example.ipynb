# If Statements

Programming often involves examining a set of conditions and deciding what action to take based on those conditions. The `if` statement will be your main tool for accomplishing this.

In this section, you will learn how to write conditional tests, which allow you to check the current state of the program against any condition of interest. 

## A Simple Example

The following example shows you how `if` tests let you respond to sepcial situations correctly. For example, lets say we're converting a list of celcius degrees to fahrenheit degrees. Before converting, it would be good to check if a given value is physical.

```python
# Several physical temperatures, and one unphysical temperature
temperatures_c = [0, 37.0, -300.0, 15e6]

for temp_c in temperatures_c:
    if temp_c < -273.15:
        print(f'{temp_c} C is not physical! The Fahrenheit temperature will not be computed')
    else:
        temp_f = 9/5*temp_c + 32
        print(f'{temp_c} = {temp_f}')
```

The loop in this example first checks if the current value of `temp_c` is greater than -273.15 C (absolute 0). If it is, it prints a message. If the value is anything greater than absolute 0, it is converted to degrees Fahrenheit and printed to the screen.
