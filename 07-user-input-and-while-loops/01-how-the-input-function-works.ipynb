# User Input and While Loops

Your programs will sometimes need to prompt the user for information. For example, you could write a program to solve the quadratic equation that prompts the user for the coefficients $a$, $b$, and $c$. In general, being able to write interactive programs is a powerful tool.

## How the input() function works

The `input()` function pauses your program and waits for the user to input a value. Python then takes this value and assigns it to a variable so that it is convenient to work with.

For example, the following program asks the users to enter a value in degrees Fahrenheit and converts it to degrees Celcius.

````python
response = input('Enter a temperature in degrees Fahrenheit: ')
degrees_f = float(response)
degrees_c = 5/9*(degrees_f - 32)
print(degrees_c)
````

Notice that we had to convert the response to number using `float()`. The `input()` function always returns a string. You are responsible for deciding what to do with it, whether that's converting it to an integer with `int()` or a floating-point number with `float()`, or something else entirely.

This isn't very useful in a Jupyter notebook, but if you copy this into a Python program (`prompt_f.py`) and executue it in the terminal, it will prompt you for a value and print the result.

### Writing Clear Prompts

Each time you use the `input()` function, you should include a clear, easy-to-follow prompt that tells the user exactly what kind of information you're looking for. Any statement that tells the user what to enter should work. Adding a colon (`:`) and a space after a prompt is often good practice to help distinguish the prompt from the value the user is entering.

```text
Enter a temperature in degrees Fahrenheit: 32
0.0
```

### The Modulo Operator

Another useful tool for working with numbers is the *modulo operator* (`%`). It divides one number by another and returns the remainder.

````python
4 % 3
````

````python
5 % 3
````

````python
6 % 3
````

The modulo operator doesn't tell you how many times the second number fits into the first numer, it only tells you what the remainder is.
