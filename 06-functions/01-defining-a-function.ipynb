# Functions

## Defining a Function

Here's a simple function names `fahrenheit_to_celcius()` that converts degrees Fahrenheit to degrees Celcius:

````python
def fahrenheit_to_celcius(degrees_f):
    """Convert degrees Fahrenheit to degrees Celcius"""
    return 5/9*(degrees_f - 32)
````

This example shows a simple structure of a function. The first line uses the keyword `def` to inform Python that you're defining a function. This is the *function definition*, which tells Python the anme of the function and what kind of information the function needs to do its job. In this case, the name of the function is `fahrenheit_to_celcius` and it needs the variable `degrees_f` to do its job. Finally, the definition ends in a colon (`:`).

Any indented lines that follow `def fahrenheit_to_celcius(degrees_f):` make up the *body* of the function. The text on the second line is a comment called a *docstring*, which describes what the function does. When Python generates documentation for the functions in your programs, it looks for a string immediately after the functions definition.

When you want to use this function, you have to call it. To *call* a function, you write the name of the function, followed by any necessary information in parentheses.

````python
fahrenheit_to_celcius(32)
````

The variable `degrees_f` in the definition of `fahrenheit_to_celcius()` is an example of a *parameter*. The value `32` in `fahrenheit_to_celcius(32)` is an example of an argument. An *argument* is a piece of information that is passed from a function call to a function.

To summarize: we passed the argument `32` to the function `fahrentheit_to_celcius()`, and the value was assigned to the parameter `degrees_f`. People (including myself) sometimes speak of arguments and parameters interchangeably, since it's usually clear based on context what they mean.

The value the function returns is called the *return value*. The `return` statement takes a value from inside a function and sends it back to the line that called the function.
