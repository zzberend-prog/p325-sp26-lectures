## Return Values

We've already seen the use of a simple return value, but in general you can return any type of variable. For example, say we wanted to calculate the coefficients $a$ and $b$ for a line passing through two $(x,y)$ coordinates. This function might look like:

```python
def line(x0, y0, x1, y1):
    """Compute the coefficients m and b for a line passing through two points"""
    m = (y1-y0)/(x1-x0)
    b = y0 - m*x0
    return m, b

m, b = line(0,0,10,10)
print(f' y = {m}x + {b}')
```

Here, we've returned two values: `m` and `b`. Notice that the values are separate by a comma (`,`) which indicates that the function is in fact returning a `tuple` but without using parentheses. Outside of the function, the syntax for assiging multiple variables is used to pick out the expected values in the returned `tuple`.

You can also return a list, for example:

```python
def odd_numbers(first, last):
    """Return a list of the odd numbers between first and last (exclusive)"""
    # Check if the first number is even; if so, increment by 1
    if first % 2 == 0:
        first += 1
    # Return list of odd numbers
    return [value for value in range(first,last,2)]

print(odd_numbers(2, 10))
```
