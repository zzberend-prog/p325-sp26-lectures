# Classes

*Object-oriented programming (OOP)* is one of the most effective approaches to writing software. In object-oriented programming:
* you write *classes* that represent real-world things and situations, and
* you create *objects* based on these classes.
When you create a class, you define the general behavior that a while category of objects can have.

When you create individual objects from the class, each object is automatically equipped with the general behavior; you can then give each object whatever unique traits you need to. For example, a string (`str`) is a Python class. All strings come with the function `strip()`, even though each one has a unique set of characters (such as `'Hello, World!'`) associated with it.

Making an object from a class is called *instantiation*, and you work with *instances* of a class.

In this section, you will write classes and create instances of those classes. You will specify the kind of information that can be stored in instances, and you'll define the actions that can be taken with these instances. You will also write classes that extend the functionality of existing classes.

## Creating and Using a Class

Let's start by writing a simple class, `Dog`, that represents a dog. Not just one dog in particular, but any dog.

What do we know about most pet dogs? Well, they have a name and an age. We also know that most dogs sit and eat coding assignments. Those two pieces of information (name and age) and those two behaviors (sit and eat homework) will go in our `Dog` class because they're common to most dogs.

This class will tell Python how to make an object representing a dog.

### Creating the Dog Class

Each instance created from the `Dog` class will store a `name` and an `age`, and we'll give each dog the ability to `sit()` and `eat_homework()`:

````python
class Dog:
    """A simple attempt to model a dog."""

    def __init__(self, name, age):
        """Initialize name and age attributes."""
        self.name = name
        self.age = age

    def sit(self):
        """Simulate a dog sitting in response to a command."""
        print(f"{self.name} is now sitting.")

    def eat_homework(self):
        """Simulate a dog eating homework."""
        print(f'Oh no! {self.name.title()} just ate my homework!!')
````

There is a lot going on here, but the more you see this structure, the more repetitive it will become. The thing to highlight first is the that we start by using the `class` keyword to define a class named `Dog`. By convention, capitalized names refer to classes in Python. There are no parentheses in the class definition because we're creating this class from scratch. Finally, we write a docstring describing what this class does.

### The `__init__()` method

A function that is part of a class is a *method*. Everything you've learned about functions applies to method's as well.

The `__init__()` method is a special method that Python runs automatically whenever we create a new instance based on the `Dog` class. **Make sure to use two underscores on each sid when defining this function.**

We defined the `__init__()` method to have three parameters:
* `self`,
* `name`, and
* `age`.

The `self` parameter is required in the method definition, and it must come first. It must be included because when Python calls this method later (to crate an instance of `Dog`), the method call will automatically pass the `self` argument.

Every method call associated with an instance automatically pass `self`, which is a reference to the instance itself; it gives the individual instance access to the attributes and methods in the class. Whenever we make an instance from the `Dog` class, we'll provide values for only the last two parameters: `name` and `age`.

The two variables defined in the body of the `__init__()` method each have the prefix `self`. Any variable prefixed with `self` is available to every method in the class, and we'll be able to access these variables through any instance created from the class. The line:
```python
self.name = name
```
takes the value associated with the parameter `name` and assigns it to the variable `name`, which is then attached to the instance being created. The same process happens with `self.age = age`.

Variables that are accessible through instances like this are called *attributes*.

### Other Methods

The `Dog` class has two other methods defined: `sit()` and `eat_homework()`. Because these methods don't need additional information to run, we define them to have one parameter, `self`. The instances we create later will have access to these methods.

For now, these methods don't do much. (They are just print statements.) Later, we'll extend methods to more realistic (and useful) situations.

### Making an Instance from a Class

Think of a class as a set of instructions for how to make an instance. For example, let's make an instance representing a specific dog:

````python
space_dog = Dog('Laika', 72)

print(f'The first dog in space was {space_dog.name}.')
print(f'She would have been {space_dog.age} years old.')
````

Here, we tell Python to create a dog whose name is `'Laika'` and whose age is `72`. When Python reads this line, it calls the `__init__()` method in `Dog` with the arguments `'Laika'` and `72`. The `__init__()` method creates an instance representing this particular dog and sets the `name` and `age` attributes. Python then returns this instance, which we assign to the variable `space_dog`.

#### Accessing Attributes

To access the attributes of an instance, you use dot notation. We access the value of `space_dog`'s attribute `name` by writing:
```python
space_dog.name
```
Dot notation is used often in Python. This syntax demonstrates how Python finds an attribute's value. Here, Python looks at the instance `space_dog` and then finds the attribute `name` associated with `space_dog`.

#### Calling Methods

After we create an instance from the class `Dog`, we can use dot notation to call any method defined in `Dog`. Let's make our dog sit and eat some homework:

````python
space_dog.sit()
space_dog.eat_homework()
````

This syntax is quite useful. When attributes and methods have been named well, we can easily infer what a block of code, even one we've never seen before, is supposed to do.

#### Creating Multiple Instances

You can create as many instances from a class as you need. Let's create a second dog:

````python
space_dog = Dog('Laika', 72)
newton_dog = Dog('Diamond', 350)

print(f'Space dog was named {space_dog.name}')
space_dog.sit()

print(f'\nNewton\'s dog was named {newton_dog.name}')
newton_dog.eat_homework()
````

Here, we've created a dog named Laika and a dog named Diamond. Each dog is a separate instance with its own set of attributes, capable of the same set of actions. As long as you give each instance a unique variable name, or it occupies a unique spot in a list of dictionary, you can make as many instances from one class as you need.
