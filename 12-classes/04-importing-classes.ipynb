# Classes

## Importing Classes

To keep your files as uncluttered as possible (and the code more readable) you'll usually want to store classes in modules and then import the class into your main program.

### Importing a Single Class

Copy the `Car` class we've been working on into a file called `car.py` in the same directory as this notebook (`12-classes/`).
Similar to how we've imported functions, we can now import the `Car class`.

```python
from car import Car

my_car = Car('ford', 'escape', 2026)
print(my_car.get_descriptive_name())

my_car.odometer_reading = 23
my_car.read_odometer()
```

By keeping the `Car` class in its own file, it's easy to use in multiple other programs while hiding (potentially) complicated logic.
Once the classes are working as you want them to, you can leave them alone and focus on the higher-level logic in your main program.
This keeps the main program file clean and easy to read.

### Storing Multiple Classes in a Module

You can store as many classes as you need in a single module, although each module should be related somehow.
For example, the classes `Battery` and `ElectricCar` are relevant and can be put in the same `car.py` module.
Now we can import them as usual. (Note: you will need to restart the kernel to pick up the two classes you just added to the module.)

```python
from car import ElectricCar

my_ev = ElectricCar('nissan', 'leaf', 2026)
print(my_ev.get_descriptive_name())
my_ev.battery.describe_battery()
```

### Importing a Module into a Module

Sometimes you will want to spread out your classes over multiple files to keep them from getting too long.
For example, many things use batteries, so storing `Car` and `ElectricCar` in one module and `Battery` in another module makes sense.
In this case, you can import the `Battery` class from a new module, `battery.py`, and everything else stays the same.

```python
from car import ElectricCar

my_ev = ElectricCar('nissan', 'leaf', 2026)
print(my_ev.get_descriptive_name())
my_ev.battery.describe_battery()
```

### Finding Your Own Workflow

On top of what we've just gone over, you can alias classes as we've seen before with functions and import multiple (or all!) classes in a module.
This gives you many options for structuring your own code.
In general, my advice is to keep your code structure simple.
Often, starting by putting everything in one file to start with and then separating things out as you get an idea of how things connect is a reasonable way to proceed.
