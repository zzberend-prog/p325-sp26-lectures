# Classes

## Working with Classes and Instances

Once you write a class, you'll spend most of your time working with instances created from that class. One of the first things we'll do is modify the attributes associated with a particular instance. This can be done directly or using methods that update attributes in a specific way.

### The Car Class

Let's write a new class representing a car. It will store information about the kind of car we're working with and it will have a method that summarizes that information.

```python
class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        """Initialize attributes to describe a car"""
        self.make = make
        self.model = model
        self.year = year

    def get_descriptive_name(self):
        """Return a neatly formatted descriptive name"""
        long_name = f'{self.year} {self.make} {self.model}'
        return long_name.title()
```

```python
my_new_car = Car('audi', 'a4', 2026)
print(my_new_car.get_descriptive_name())
```

In this simple class, we've defined a car with the following attributes:
- `make`
- `model`
- `year`
This are all set by the `__init()__` method when the class is instantiated. As written, this method requires that all three values be specified during instantiation.

Currently, the class has a single method:
- `get_descriptive_name()`

### Setting a Default Value for an Attribute

When an instance is created, attributes can be defined without being passed as parameters. Let's add an attribute called `odometer_reading` that always starts with a value of `0`. We'll also add a method for printing the current value of this new variable.

```python
class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        """Initialize attributes to describe a car"""
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # Default value

    def get_descriptive_name(self):
        """Return a neatly formatted descriptive name"""
        long_name = f'{self.year} {self.make} {self.model}'
        return long_name.title()

    def read_odometer(self):
        """Print a statement showing the car's mileage"""
        print(f'This car has {self.odometer_reading} miles on it.')
```

```python
my_new_car = Car('audi', 'a4', 2026)
print(my_new_car.get_descriptive_name())
my_new_car.read_odometer()
```

This time, when Python calls the `__init__()` method to create a new instance, it also creates an attribute called `odometer_reading` and sets its initial value to `0`.

### Modifying Attribute Values

There are three ways to change an attribute's value:
1. Change it directly through the instance
2. Set the value through a method
3. Increment the value through a method

#### Modifying an Attribute's Value Directly

The simplest way to modify the value of an attribute is to access the attribute directly through an instance:

```python
my_new_car = Car('audi', 'a4', 2026)
my_new_car.read_odometer()

my_new_car.odometer_reading = 42
my_new_car.read_odometer()
```

#### Modifying an Attribute's Value Through a Method

It's often helpful to have methods that update certain attributes for you. Instead of accessing the attribute directly, you pass the new value to a method that handles the updating internally.

For example:

```python
class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        """Initialize attributes to describe a car"""
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # Default value

    def get_descriptive_name(self):
        """Return a neatly formatted descriptive name"""
        long_name = f'{self.year} {self.make} {self.model}'
        return long_name.title()

    def read_odometer(self):
        """Print a statement showing the car's mileage"""
        print(f'This car has {self.odometer_reading} miles on it.')

    def update_odometer(self, mileage): # new method to update mileage
        """Set the odometer reading to the given value"""
        self.odometer_reading = mileage
```

This is often used if an attributes value needs to be validated before changing it, or if it's a computed value. In our case, the mileage should never decrease, so we can check this within the method:

```python
class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        """Initialize attributes to describe a car"""
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # Default value

    def get_descriptive_name(self):
        """Return a neatly formatted descriptive name"""
        long_name = f'{self.year} {self.make} {self.model}'
        return long_name.title()

    def read_odometer(self):
        """Print a statement showing the car's mileage"""
        print(f'This car has {self.odometer_reading} miles on it.')

    def update_odometer(self, mileage): # new method to check and update mileage
        """Set the odometer reading to the given value"""
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
```

```python
my_new_car = Car('audi', 'a4', 2026)
my_new_car.read_odometer()

my_new_car.update_odometer(42)
my_new_car.read_odometer()

my_new_car.update_odometer(22)
my_new_car.read_odometer()
```

#### Incrementing an Attribute's Value Through a Method

Sometimes you'll want to increment an attribute's value by a certain amount, rather than setting an entirely new value. Say we took the car on a road trip and drove 100 miles. Instead of keeping track of where we started and what the new total mileage is, we could instead add 100 miles to the current value:

```python
class Car:
    """A simple attempt to represent a car."""

    def __init__(self, make, model, year):
        """Initialize attributes to describe a car"""
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0  # Default value

    def get_descriptive_name(self):
        """Return a neatly formatted descriptive name"""
        long_name = f'{self.year} {self.make} {self.model}'
        return long_name.title()

    def read_odometer(self):
        """Print a statement showing the car's mileage"""
        print(f'This car has {self.odometer_reading} miles on it.')

    def update_odometer(self, mileage):
        """Set the odometer reading to the given value"""
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")

    def increment_odometer(self, miles): # New method to increment mileage
        """Add the given amount to the odometer reading."""
        if miles >= 0:
            self.odometer_reading += miles
        else:
            print("Driving in reverse doesn't decrease the mileage!")
```

```python
my_new_car = Car('audi', 'a4', 2026)
my_new_car.read_odometer()

my_new_car.update_odometer(42)
my_new_car.read_odometer()

my_new_car.increment_odometer(100)
my_new_car.read_odometer()
```
