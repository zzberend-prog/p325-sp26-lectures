# Dictionaries

Dictionaries offer a convenient and flexible way to connect peices of related information. In particular, they let you store any two kinds of information that can be matched up, such as a label `'x_points'` and a list `[0, 1, 2]`.

## A Simple Dictionary

We're all hungry, but if we wanted to order pizza, we'd need to keep track of who likes what pizza. A simple dictionary to accomplish this might be:

```python
pizza_preference = {'name': 'Chris', 'pizza': 'Margherita'}

print(pizza_preference)
print(f'Name: {pizza_preference['name']}')
print(f'Pizza: {pizza_preference['pizza']}')
```

The dictionary stores a name and a pizza preference. The last two lines access and display the information.
