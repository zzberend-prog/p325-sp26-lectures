### Examples and Practice
The symbol lambda is often used in physics equations, but it is also a Python keyword. In cases like this, it is common to use an underscore at the end:

```python
lambda_= 5
print(lambda_)
```

Take the following kinematic calculation of the motion of a ball with an initial speed $v_0$ due to the acceleration of earths gravity $g=9.81$:
$$
y(t) = v_0 t - \frac{1}{2}g t^2
$$
If $v_0 = 5$ m/s, at $t=0.6$ s it will have a height of:

```python
0.5*0.6 - 0.5*9.81*0.6**2
```

If you want to compute the height at a different time, you would need to change the value in two places. An alternative is to use variables:

```python
v0 = 5
g = 9.81
t = 0.6
y = v0*t - 0.5*g*t**2
print(y)
```

Try executing the above, then change the time to 0.1 s and try executing again. Much easier!

Practice naming variables by updating the code below. For example, you could replace `v0` by `initial_velocity`.

```python
initial_velocity = 5
g = 9.81
t = 0.6
y = initial_velocity*t - 0.5*g*t**2
print(y)
```

In this case, v0 and g are common enough that using short values may be clear enough given the context. In this case, the shorter variable `v0` might make the equation on the fourth line easier to parse.

However, for less common variable names, for example `air_resistance`, you will want to spell them out.
